# 문제 1

```python
from ortools.linear_solver import pywraplp

solver = pywraplp.Solver.CreateSolver('SCIP')

# 의사결정변수
Saudi = solver.NumVar(0, solver.infinity(), 'SaudiOil')
Venezuela = solver.NumVar(0, solver.infinity(), 'VenezuelaOil')


# 제약조건
solver.Add(Saudi <= 9000) # SaudiOil 최대 구입 가능량
solver.Add(Venezuela <= 6000) # VenezuelaOil 최대 구입 가능량
solver.Add(0.3 * Saudi + 0.4 * Venezuela >= 2000) # 최소 가솔린 공급량
solver.Add(0.4 * Saudi + 0.2 * Venezuela >= 1500) # 최소 제트유 공급량
solver.Add(0.2 * Saudi + 0.3 * Venezuela >= 500) # 최소 제트유 공급량

# 목적함수
solver.Minimize(60 * Saudi + 55 * Venezuela)

# 최적화 문제 풀이
status = solver.Solve()

# 출력
if status == pywraplp.Solver.OPTIMAL:
    print('OPTIMAL:')
    print(f'목적함수값 = ', solver.Objective().Value())
    print(f'SaudiOil = %.1f' % Saudi.solution_value())
    print(f'VenezuelaOil = %.1f' % Venezuela.solution_value())
else:    
    print('The problem does not have an optimal solution.')
```

추가> nicer version

```python
from ortools.linear_solver import pywraplp

solver = pywraplp.Solver.CreateSolver('SCIP')

data = {}
data["constraint_coeffs"] = [
    [0.3,0.4],
    [0.4,0.2],
    [0.2,0.3]
]

data["supply"] = [2000, 1500, 500]
data["prices"] = [60, 55]
data["num_vars"] = 2
data["num_constraints"] = 3

infinity = solver.infinity()

# 의사결정변수
x = {}
for j in range(data["num_vars"]):
    x[j] = solver.NumVar(0, solver.infinity(), 'x[%i]' % j)

# 제약식
for i in range(data["num_constraints"]):
    constraint_expr = [data["constraint_coeffs"][i][j] * x[j] for j in range(data["num_vars"])]
    solver.Add(sum(constraint_expr) >= data["supply"][i])

# 목적함수
objective = solver.Objective()
obj_expr = [data["prices"][j] * x[j] for j in range(data["num_vars"])]
solver.Minimize(solver.Sum(obj_expr))

print(f"Solving with {solver.SolverVersion()}")
status = solver.Solve()

if status == pywraplp.Solver.OPTIMAL:
    print('Opjective value=', solver.Objective().Value())
    for j in range(data["num_vars"]):
        print(x[j].name(), "=", x[j].solution_value())
else:
    print('The problem does not have an optimal solution.')
```

# 문제 2

```python
from ortools.linear_solver import pywraplp

solver = pywraplp.Solver.CreateSolver('SCIP')

# 의사결정변수
stone1 = solver.NumVar(0, solver.infinity(), 'stone1')
stone2 = solver.NumVar(0, solver.infinity(), 'stone2')


# 제약조건
solver.Add(0.002 * stone1 + 0.003 * stone2 >= 8) # 최소 가솔린 공급량
solver.Add(0.0015 * stone1 + 0.0025 * stone2 >= 6) # 최소 제트유 공급량
solver.Add(0.002 * stone1 + 0.001 * stone2 >= 4) # 최소 제트유 공급량

# 목적함수
solver.Minimize(10 * stone1 + 15 * stone2)

# 최적화 문제 풀이
status = solver.Solve()

# 출력
if status == pywraplp.Solver.OPTIMAL:
    print('OPTIMAL:')
    print(f'목적함수값 = ', solver.Objective().Value(), '만원')
    print(f'원석1 = %.1f톤' % stone1.solution_value())
    print(f'원석2 = %.1f톤' % stone2.solution_value())
else:    
    print('The problem does not have an optimal solution.')
```

# 문제 3

```python
from ortools.linear_solver import pywraplp

solver = pywraplp.Solver.CreateSolver('SCIP')

# 의사결정변수
stroller = solver.NumVar(0, solver.infinity(), 'stroller')
babywalker = solver.NumVar(0, solver.infinity(), 'babywalker')
bycicle = solver.NumVar(0, solver.infinity(), 'bycicle')


# 제약조건
solver.Add(8 * stroller + 3 * babywalker + 3 * bycicle <= 240) # 최소 가솔린 공급량
solver.Add(4 * stroller + 4 * babywalker + 0 * bycicle <= 200) # 최소 제트유 공급량
solver.Add(4 * stroller + 0 * babywalker + 2 * bycicle <= 100) # 최소 제트유 공급량

# 기계1 여유량 = 240 - (8 * stroller + 3 * babywalker + 3 * bycicle)
# 기계3 여유량 = 100 - (4 * stroller + 0 * babywalker + 2 * bycicle)

# 목적함수
# 판매수익 : 30 * stroller + 20 * babywalker + 16 * bycicle
# 임대수익 : 3.5 * (기계1 여유량) + 3 * (기계3 여유량)
solver.Maximize(30 * stroller + 20 * babywalker + 16 * bycicle + 3.5 * (240 - (8 * stroller + 3 * babywalker + 3 * bycicle)) + 3 * (100 - (4 * stroller + 0 * babywalker + 2 * bycicle)))

# 최적화 문제 풀이
status = solver.Solve()

# 출력
if status == pywraplp.Solver.OPTIMAL:
    print('OPTIMAL:')
    print(f'목적함수값 = ', solver.Objective().Value(), '만원')
    print(f'스털러 = %.1f대' % stroller.solution_value())
    print(f'베이비워커 = %.1f대' % babywalker.solution_value())
    print(f'자전거 = %.1f대' % bycicle.solution_value())
else:    
    print('The problem does not have an optimal solution.')
```
